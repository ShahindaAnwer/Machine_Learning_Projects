```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

#**Data Analysis and Preproessing**

```python
bcdata = sklearn.datasets.load_breast_cancer()
```

```python
# putting the data to a dataframe
data = pd.DataFrame(bcdata.data, columns = bcdata.feature_names)
```

#**Separating the Data**

```python
X = data.drop(columns='label', axis=1)
Y = data['label']
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

#**Model training and Evaluation**

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
trainpred = model.predict(X_train)
trainpred
```

```python
train_accuracy = accuracy_score(trainpred, Y_train)
train_accuracy
```

```python
testpred = model.predict(X_test)
testpred
```

```python
test_accuracy = accuracy_score(testpred, Y_test)
test_accuracy
```
