# **Data Collection**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('/content/creditcardfraud.csv')
```

```python
# legit vs fraud
data['Class'].value_counts()
```

Data is highly unbalanced

0 --> Normal Transaction

1 --> fraudulent transaction

```python
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
```

```python
legit.shape
```

```python
fraud.shape
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# comparing values
data.groupby('Class').mean()
```

Bulding a sample dataser (undersampling) with similar distribution of normal transactions and Fraudulent Transactions, so that both legit and fraud would have equal amount of values

```python
legit_sample = legit.sample(n=492)
```

```python
newdata = pd.concat([legit_sample, fraud], axis = 0)
```

```python
newdata.head()
```

```python
newdata.describe()
```

```python
newdata['Class'].value_counts()
```

```python
newdata.groupby('Class').mean()
```

# **Training the Model**

```python
X = newdata.drop('Class', axis = 1)
Y = newdata['Class']
```

```python
print(X)
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
model.predict(X_train)
```

# **Model Evaluation**

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
